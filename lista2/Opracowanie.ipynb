```python
%matplotlib inline

import csv
import collections

import matplotlib.pyplot as plt
import numpy as np

elock = []
padlock = []


with open('wpisy_rowerowe.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        dif = int(row['roznica'])
        elock.append(dif) if row['group'] == '0' else padlock.append(dif)
```

```python
hist_elock, bins, patches = plt.hist(elock, facecolor='b', bins=4)
plt.title('Rozklad rowerow na zamku elektrycznym')
plt.xlabel('Odleglosc')
plt.ylabel('Liczba rowerow')
plt.show()
```

```python
hist_padlock, bins, patches = plt.hist(padlock, facecolor='b', bins=4)
plt.title('Rozklad rowerow na zamku elektrycznym')
plt.xlabel('Odleglosc')
plt.ylabel('Liczba rowerow')
plt.show()
```
